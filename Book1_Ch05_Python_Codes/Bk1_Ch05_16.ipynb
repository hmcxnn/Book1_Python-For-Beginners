```python
# 使用大括号创建字典
person = {'name': 'James', 'age': 88, 'gender': 'male'}
```

```python
# 使用 dict() 函数创建字典
fruits = dict(apple=88, banana=888, cherry=8888)
```

```python
# 访问字典中的值
print(person['name']) 
print(fruits['cherry'])  
print(fruits)
print(person)
```

```python
# 修改字典中的值
person['age'] = 28
print(person)  
fruits['apple'] = 99
print(fruits)
```

```python
# 添加键值对
person['city'] = 'Toronto'
print(person)
```

```python
# 删除键值对
del person['city']
print(person)  
del fruits['apple']
print(fruits)


```

```python
#del 不提供撤销,如果需要可使用pop或者创建副本
person = {'name': 'Tom', 'age': 25, 'city': 'Beijing'}
deleted_value = person.pop('city')  # 删除并返回值 'Beijing'
print(person)  # {'name': 'Tom', 'age': 25}
print(deleted_value)  # 'Beijing'
# 可以重新添加回来
person['city'] = deleted_value
```

```python
# 获取键、值、键值对列表
print(person.keys())  
print(person.values())  
print(person.items())  
```
