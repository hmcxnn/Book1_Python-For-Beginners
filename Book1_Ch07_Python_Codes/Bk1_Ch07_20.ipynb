# 利用一层列表生成式完成矩阵转置
Book_1《编程不难》 | 鸢尾花书：从加减乘除到机器学习 

```python
def transpose(matrix):
    transposed = []
    rows = len(matrix)
    cols = len(matrix[0])

    for j in range(cols):
        transposed_row = [matrix[i][j] 
                          for i in range(rows)]
        transposed.append(transposed_row)

    return transposed
```

```python
# 示例用法
A = [[1, 2, 3],
     [4, 5, 6]]
```

```python
# 调用自定义函数
B = transpose(A)
B
```

```python
import numpy as np
B = np.array(A).T
print(B)
```
