# 采用 * 和 ** 拆包
Book_1《编程不难》 | 鸢尾花书：从加减乘除到机器学习 

```python
# 使用一个星号
complex_list = [3, 4]
complex(*complex_list)
```

```python
# 使用两个星号
complex_dict = {'real': 3, 'imag': 4}
complex(**complex_dict)
```

```python
def func(**kwargs):
    return complex(**kwargs)
a={'real':1,'imag':2}
print(func(**a))
```
