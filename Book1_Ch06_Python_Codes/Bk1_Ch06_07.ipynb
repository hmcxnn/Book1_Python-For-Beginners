```python
import copy
a = [1, 2, 3]
b = a  
```

```python
# 视图 b 引用 a 的内存地址
c = [1, 2, 3]
d = a.copy()
```

```python
print(a is b)        
# 输出 True，因为 a 和 b 引用同一个内存地址
```

```python
print(a is not c)    
# 输出 True，因为 a 和 c 引用不同的内存地址
```

```python
print(a == c)       
# 输出 True，因为 a 和 c 的值相等
```

```python
print(a is not d)   
# 输出 True，因为 a 和 d 引用不同的内存地址
```

```python
print(a == d)      
# 输出 True，因为 a 和 d 的值相等
```

```python
a_2_layers = [1, 2, [3, 4]]
a_2_layers
```

```python
d_2_layers = a_2_layers.copy() 
d_2_layers
```

```python
e_2_layers = copy.deepcopy(a_2_layers) 
e_2_layers
```

```python
print(a_2_layers is d_2_layers)# is 检测内存地址
print(a_2_layers[2] is d_2_layers[2]) # 请特别关注! 应为是浅复制,所以嵌套列表是同一个地址
print(a_2_layers[2] is e_2_layers[2])#深复制,不是同一地址
```

```python
print(a_2_layers is e_2_layers)
print(a_2_layers[2] is e_2_layers[2])
```

```python
a_2_layers[2][1]=5
print(a_2_layers)
print(d_2_layers)#浅复制
print(e_2_layers)#深复制
```
