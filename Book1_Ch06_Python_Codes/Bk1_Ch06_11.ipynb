# 混合两个一元高斯分布随机数
Book_1《编程不难》 | 鸢尾花书：从加减乘除到机器学习 

```python
import random
import statistics
import matplotlib.pyplot as plt
```

```python
random.seed(0) # 方便复刻结果
# 生成300个服从N(0, 1**2)的随机数
mean1, std1, size1 = 0, 1, 300
data1 = [random.gauss(mean1, std1) for _ in range(size1)]
```

```python
# 生成700个服从N(10, 5**2)的随机数 
mean2, std2, size2 = 10, 5, 700
data2 = [random.gauss(mean2, std2) for _ in range(size2)]
```

```python
# 将两组随机数混合
mixed_data = data1 + data2
mean_loc = statistics.mean(mixed_data)
std_loc  = statistics.stdev(mixed_data)
```

```python
# 可视化混合随机数的直方图
import matplotlib.pyplot as plt
random.shuffle(mixed_data)
plt.figure(figsize=(6, 4), dpi=150)
plt.hist(mixed_data, bins=50, density=True, alpha=0.7, color='steelblue', edgecolor='white', label='Mixed')
plt.axvline(mean_loc, color='red', linestyle='--', label=f'mean={mean_loc:.2f}')
plt.axvline(mean_loc - std_loc, color='orange', linestyle=':', linewidth=1)
plt.axvline(mean_loc + std_loc, color='orange', linestyle=':', linewidth=1)
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram of Mixed Data')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
