# 比较运算符
Book_1《编程不难》 | 鸢尾花书：从加减乘除到机器学习 

```python
x = 5
y = 3
```

```python
print(x == y)    # False
print(x == 5)    # True
print(x != y)    # True
print(x != 5)    # False
print(x != 5.0)  # False
```

```python
n="hello"
m="hello"
print(n==m)
```

```python
n="hello"
m="yes"
print(n==m)
```

```python
print(x > y)   # True
print(x > 10)  # False
print(x >= y)  # True
print(x >= 5)  # True
```

```python
print(x < y)   # False
print(x < 10)  # True
print(x <= y)  # False
print(x <= 5)  # True
```

```python
print("10"<"2")#按照字典序比较,比较第一个字符：'1' vs '2',看 Unicode 编码值：ord('1') = 49,ord('2') = 50
```

```python
print(5=="5")
print(5<"5")#不同类型不能比较
```
